```python
# prompt: create columns type_CASH_OUT
# type_DEBIT
# type_PAYMENT
# type_TRANSFER

import pandas as pd
import numpy as np
df=pd.read_csv('Raw_Dataset_for_Online_Payment.csv')
df.info()
df.head()
df.isnull().sum()

# Create new columns for each transaction type
df['type_CASH_OUT'] = (df['type'] == 'CASH_OUT').astype(int)
df['type_DEBIT'] = (df['type'] == 'DEBIT').astype(int)
df['type_PAYMENT'] = (df['type'] == 'PAYMENT').astype(int)
df['type_TRANSFER'] = (df['type'] == 'TRANSFER').astype(int)
```

```python
df.head()
```

```python
df.drop('type',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(columns=['step','nameOrig','nameDest','isFlaggedFraud'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['oldbalanceDest','newbalanceDest'],axis=1,inplace=True)
```

```python
# prompt: change the isFraud column to the last in the dataframe

cols = list(df.columns)
cols.remove('isFraud')
cols.append('isFraud')
df = df[cols]
```

```python
df.head()
```

```python
df.to_csv('Cleaned_Dataset_for_Online_Payment.csv',index=False)
```
